# Tugas List Nilai

Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:
1. Tampilkan Angka Ganjil dari ke 250 nilai tersebut
2. Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
3. Tampilkan standar deviasi dari sekumpulan nilai tersebut
4. Tampilkan nilai maksimum dan minimum
5. Tampilkan modus dari sekumpulan nilai tersebut

```python
import random
from math import sqrt

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print("==== Nilai Tugas ====")
print(nilaiTugas)
print()


# DISPLAY:
#=========== Odd Value =============#
nilaiGanjil = []
for i in nilaiTugas:
    if i%2==1: # determine the odd value using modulo division
        nilaiGanjil.append(i)
        
print("==== Nilai Ganjil ====")
print(nilaiGanjil)
print()


#=========== Value > 70 ============#
nilaiLebih70 = []
for i in nilaiTugas:
    if i > 70: # determine the value that greater than 70
        nilaiLebih70.append(i)
        
print("==== Nilai Lebih dari 70 ====")
print(nilaiLebih70)
print("--- Total jumlah nilai yang lebih dari 70 = ", len(nilaiLebih70))
print()


#======= Standard Deviation ========#
def mean(lst):
    ## function to calculate mean
    sum = 0
    for i in lst:
        sum += lst[i]
    return (sum / len(lst))

def stddev(lst):
    ## function to calculate standart deviation
    sum = 0
    mn = mean(lst)
    for i in lst:
        sum += pow((lst[i]-mn),2)
    return sqrt(sum/len(lst)-1)

print("==== Nilai Standar Deviasi = ", stddev(nilaiTugas))
print()


#============ Maximum ==============#
### print(max(nilaiTugas))
## creating a max function from scratch
def maximum(lst):
   Max = 0
   for i in lst:
       if i > Max:
           Max = i
   return Max

print("==== Nilai Maksimum = ", maximum(nilaiTugas))
print()


#============ Minimum ==============#
### print(min(nilaiTugas))
## creating a min function from scratch
def minimum(lst):
   Min = 0
   for i in lst:
       if i < Min:
           Min = i
   return Min

print("==== Nilai Minimum = ", minimum(nilaiTugas))
print()


#============= Mode ================#
### from statistics import mode
### print (mode(nilaiTugas))
## creating a mode function from scratch
def mode(lst):
    newList = []
    for i in lst:
        newList.append(lst.count(lst[i]))
    m = maximum(newList)

    ml = [x for x in lst if lst.count(x)== m ] # find most frequent values
    mode = []
    for x in ml: # remove duplicates of mode
        if x not in mode:
            mode.append(x)
    return mode

print("==== Nilai Modus = ", mode(nilaiTugas))
print()
```
