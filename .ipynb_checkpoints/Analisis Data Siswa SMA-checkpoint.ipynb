# Analisis Data Siswa SMA
Dataset awal berisi presentase siswa SMA menurut tempat tinggal, di setiap provinsi di Indonesia yang didapatkan dari [data.go.id](https://data.go.id/dataset/presentase-siswa-menurut-tempat-tinggal). 
- Data bersumber dari Pusat Data dan Statistik Pendidikan dan Kebudayaan Indonesia
- Data terverifikasi pada tanggal 23 Oktober 2018


## 1. Total Siswa SMA berdasarkan lokasi tempat tinggal
Grafik berikut menunjukkan total siswa SMA berdasarkan lokasi tempat tinggalnya, serta menampilkan beberapa tabel  antara siswa laki-laki dan perempuan baik di sekolah negeri maupun disekolah swasta serta jumalah total siswa disetiap daerah

```python
import pandas as pd

data_siswa = pd.read_csv('data/TempatTinggalSiswa_SMA_Indonesia.csv', delimiter = ',', encoding = "ISO-8859-1")

data_siswa # display data
```

## 2. Total Siswa SMA berdasarkan Provinsi
Tabel berikut menunjukkan total jumlah siswa SMA berdasarkan seluruh provinsi di Indonesia

```python
data_siswa[['nama_provinsi','total']]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

plt.rcdefaults()
fig, ax = plt.subplots()

# set data
siswa = data_siswa['nama_provinsi']
y_pos = np.arange(len(siswa))
total = data_siswa['total']

ax.barh(y_pos, total, align='center', color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(siswa, fontsize=8)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Siswa SMA')
ax.set_title('Jumlah Siswa SMA Berdasarkan Provinsi')

plt.show()

```

## 3. Provinsi dengan siswa paling banyak
berikut 5 provinsi teratas yang memiliki jumlah siswa terbanyak


```python
data_siswa[['nama_provinsi','total']].sort_values(by='total', ascending=False).head()
```

## 4. Provinsi dengan siswa paling sedikit
berikut 5 provinsi teratas yang memiliki jumlah siswa paling sedikit


```python
data_siswa[['nama_provinsi','total']].sort_values(by='total', ascending=True).head()
```

## 4. DKI Jakarta
### Perbandingan jumlah siswa SMA Negeri Dan Swasta di DKI Jakarta
Berikut Perbandingan jumlah siswa SMA di sekolah negeri dan sekolah swasta di DKI Jakarta

```python
dki_negeri = data_siswa['siswa_putra_sekolah_negeri'][0] + data_siswa['siswa_putri_sekolah_negeri'][0]
dki_swasta = data_siswa['siswa_putra_sekolah_swasta'][0] + data_siswa['siswa_putri_sekolah_swasta'][0]

print("Jumlah Siswa SMA Negeri di Jakarta = ", dki_negeri)
print("Jumlah Siswa SMA Swasta di Jakarta = ", dki_swasta)
```

### Persentase perbandingan siswa SMA Negeri dan Swasta di DKI Jakarta
Berikut Persentase Perbandingan jumlah siswa SMA di sekolah negeri dan sekolah swasta di DKI Jakarta

```python
# Pie chart
labels = 'Siswa SMA Negeri', 'Siswa SMA Swasta'
sizes = [dki_negeri, dki_swasta]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Persentase perbandingan siswa SMA Negeri dan Swasta di DKI Jakarta')

plt.show()
```

### Persentase perbandingan siswa SMA Putra dan Putri di DKI Jakarta
Berikut Persentase Perbandingan jumlah siswa SMA Putra dan Putri di DKI Jakarta

```python
dki_putra = data_siswa['siswa_putra_sekolah_negeri'][0] + data_siswa['siswa_putra_sekolah_swasta'][0]
dki_putri = data_siswa['siswa_putri_sekolah_negeri'][0] + data_siswa['siswa_putri_sekolah_swasta'][0]

print("Jumlah Siswa SMA Putra di Jakarta = ", dki_putra)
print("Jumlah Siswa SMA Putri di Jakarta = ", dki_putri)
```

```python
# Pie chart
labels = 'Putra', 'Putri'
sizes = [dki_putra, dki_putri]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Persentase perbandingan siswa SMA Putra dan Putri di DKI Jakarta')

plt.show()
```
